```python
CONVERT_ALL = True
destdir = pwd_path[0]+'/content/target_kw/'
pbar = sorted(os.listdir(audio_dir)) # for viewing progress
print(pbar)
for word in pbar:
    print("word name:" , word)
    # pbar.set_description(f"Converting {language}/{word}")
    destdir_word = destdir + word
    if not os.path.exists(destdir_word):
        os.makedirs(destdir_word)
    
    wavs_dir = audio_dir + word
    samples = list(Path(wavs_dir).rglob("*.wav"))
    samples.sort()
    print("number of wav files:", len(samples))
    #if not CONVERT_ALL and len(samples) > N_SAMPLES_TO_CONVERT: 
    #    samples = rng.choice(samples, N_SAMPLES_TO_CONVERT, replace=False)
    if len(samples)> 0:
        for wav_path in samples:
            #wav_path = destdir / (opus_path.stem + ".wav")
            dest_file = os.path.join(destdir_word,  (wav_path.stem + ".wav"))
            print(dest_file )
            print(wav_path)
            shutil.copy(wav_path, dest_file) # shutil.copy(origin, target)
            cmd = ["ffmpeg", "-i", wav_path, "-c:a pcm_s16le", dest_file] # see markdown cell above for reference
            subprocess.run(cmd)
```
